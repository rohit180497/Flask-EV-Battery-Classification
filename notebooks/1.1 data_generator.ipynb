```python
import pandas as pd
import numpy as np

def add_random_missing_values(df, missing_ratio=0.4, max_missing_percentage=0.35):
    """
    Add random missing values to a DataFrame with varying missing percentages.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    missing_ratio (float): The ratio of columns to add missing values to.
    max_missing_percentage (float): The maximum percentage of missing values to add to each column.

    Returns:
    pd.DataFrame: A new DataFrame with missing values added.
    """

    # Create a copy of the input DataFrame to avoid modifying the original
    new_df = df.copy()

    # Determine the number of columns to add missing values to
    num_columns_to_add_missing = int(len(new_df.columns) * missing_ratio)

    # Randomly select columns to add missing values to
    columns_to_add_missing = np.random.choice(new_df.columns, num_columns_to_add_missing, replace=False)

    for column in columns_to_add_missing:
        # Determine the maximum percentage of missing values to add to this column
        max_missing_percentage_column = np.random.uniform(0, max_missing_percentage)

        # Randomly generate missing values for this column
        missing_mask = np.random.rand(len(new_df)) < max_missing_percentage_column

        # Set the selected indices to NaN in the column
        new_df.loc[missing_mask, column] = np.nan

    return new_df
```

```python
import pandas as pd
import numpy as np

def add_random_outliers(df, outlier_ratio=0.4, max_outlier_percentage=0.1, outlier_range=(-100, 100)):
    """
    Add random outliers to a DataFrame in selected columns.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    outlier_ratio (float): The ratio of columns to add outliers to.
    max_outlier_percentage (float): The maximum percentage of outliers to add to each column.
    outlier_range (tuple): The range within which to generate outliers (min, max).

    Returns:
    pd.DataFrame: A new DataFrame with outliers added.
    """

    # Create a copy of the input DataFrame to avoid modifying the original
    new_df = df.copy()

    # Determine the number of columns to add outliers to
    num_columns_to_add_outliers = int(len(new_df.columns) * outlier_ratio)

    # Randomly select columns to add outliers to
    columns_to_add_outliers = np.random.choice(new_df.columns, num_columns_to_add_outliers, replace=False)

    for column in columns_to_add_outliers:
        # Determine the maximum percentage of outliers to add to this column
        max_outlier_percentage_column = np.random.uniform(0, max_outlier_percentage)

        # Calculate the number of outliers to add
        num_outliers = int(len(new_df) * max_outlier_percentage_column)

        # Generate random outlier values within the specified range
        outliers = np.random.uniform(outlier_range[0], outlier_range[1], num_outliers)

        # Randomly select indices to replace with outliers
        indices_to_replace = np.random.choice(len(new_df), num_outliers, replace=False)

        # Set the selected indices to outliers in the column
        new_df.loc[indices_to_replace, column] = outliers

    return new_df
```
